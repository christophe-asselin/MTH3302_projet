### Gestion des données manquantes
- #### en calculant la moyenne des autres stations
 * Lorsque l'on ne trouve pas une donnée de précipitation pour une certaine station, on fait la moyenne des précipitations pour les autres station de notre échantillon.
 * Une amélioration qui aurait pu être apportée à cette technique aurait été de pondérer la moyenne par la distance entre la station aux données manquantes et la station utilisé dans la moyenne. 
- #### en utilisant une regression Ridge pour prédire une valeur
 * Une autre façon de trouver des valeurs pour les précipitations manquantes utilisée a été d'effectuer une régression Ridge pour prédire des valeurs de précipitations. Cette méthode a été testée, mais fonctionnait moins bien que de faire la moyenne. Nous avons utlisé la librairie Julia MultivariateStats pour effectuer la regression. Le code suivant à été utilisé:
 
```julia
stations = [Symbol("McTavish"), Symbol("Bellevue"), Symbol("Assomption"), Symbol("Trudeau"), Symbol("StHubert")]

precipitation = CSV.read("data/precipitations.csv",missingstring="-99999")
rename!(precipitation, Symbol("St-Hubert")=>:StHubert)
first(precipitation,5)
precipitation = filter(row -> month(row.date) > 4, data) 
precipitation = filter(row -> month(row.date) < 11, data)
dropmissing!(precipitation)

for i in 1:size(data,1)
    missingValues = Symbol[]
    for station in stations
        if ismissing(data[i, station])
            push!(missingValues, station)
        end
    end
    if size(missingValues, 1) == 5
        for missingValue in missingValues
            data[i, missingValue] = 0
        end
    elseif size(missingValues, 1) > 0
        for missingValue in missingValues
            X = convert(Matrix{Float64}, precipitation[:, Not(vcat(Symbol[:date, :heure], missingValues))])
            X̃ = convert(Array{Float64,1}, data[i, Not(vcat(Symbol[:date, :heure], missingValues))])'
            Y = convert(Array{Float64}, precipitation[:, missingValue])
            β = ridge(X, Y, 0; bias=false)
            Ŷ = X̃*β
            data[i, missingValue] = round(Ŷ)
        end
    end

end
first(data,5)
```
#### Gestion des valeurs manquantes avec la moyenne des autres stations

# 2. Analyse exploratoire

Cette section consitue une analyse exploratoire superficielle permettant de voir s'il existe un lien entre les précipitations et les surverses.

Prenons arbitrairement l'ouvrage de débordement près du Bota-Bota (4350-01D). La station météorologique la plus proche est McTavish. Prenons huit variables explicatives :
- la somme journalière des précipitations
- le taux horaire maximum journalier de précipitations
- l'intensité journalière des précipitations (somme / nombre d'heures de précipitation)
- le nombre d'heures de précipitation journalier
- la plus grande période de précipitation journalière
- la somme des précipitations par blocs de trois jours
- le maximum journalier de précipitations en deux heures
- le maximum journalier de précipitations en trois heures

#### Calcul de la quantité journalière de précipitations pour chacune des stations météorologiques

Comme indiqué dans le nom, on compte le nombre de mm de pluie accumulé sur la journée

````julia
pcp_sum = by(data, :date,  McTavish = :McTavish => sum, Bellevue = :Bellevue => sum, 
   Assomption = :Assomption => sum, Trudeau = :Trudeau => sum, StHubert = :StHubert => sum)
first(pcp_sum, 5)
````

#### Extraction du taux horaire journalier maximum des précipitations pour chacune des stations météorologiques

On cherche à trouver la quantité maximale de pluie tombée en une heure dans la journée.

````julia
pcp_max = by(data, :date,  McTavish = :McTavish => maximum, Bellevue = :Bellevue => maximum, 
   Assomption = :Assomption => maximum, Trudeau = :Trudeau => maximum, StHubert = :StHubert => maximum)
first(pcp_max, 5)
````

#### Calcul de l'intensité journalière de précipitation

On cherche à savoir l'intensité d'une précipitations en une journée. Pour calculer :
1. On calcule la somme *s* des précipitation de la journée.
2. On calcule le nombre d'heure *t* où il y eu de la pluie.

Puis on calcule *intensité = s / t*

````julia
function compute_intensity(itr)
    pcp_sum = sum(itr)
    pcp_hr_count = count(i -> ismissing(i) ? false : i > 0, itr)
    return pcp_hr_count == 0 ? 0 : pcp_sum / pcp_hr_count
end

pcp_intensity = by(data, :date,
                    McTavish = :McTavish => x -> compute_intensity(x),
                    Bellevue = :Bellevue => x -> compute_intensity(x),
                    Assomption = :Assomption => x -> compute_intensity(x),
                    Trudeau = :Trudeau => x -> compute_intensity(x),
                    StHubert = :StHubert => x -> compute_intensity(x))

first(pcp_intensity, 5)
````

#### Calcul du nombre d'heures de pluie journalier

On cherche à compter combien d'heures il a plu dans la journée.

````julia
function count_pcp_hours(itr)
    count(i -> ismissing(i) ? false : i > 0, itr)
end

pcp_hours = by(data, :date,
                    McTavish = :McTavish => x -> count_pcp_hours(x),
                    Bellevue = :Bellevue => x -> count_pcp_hours(x),
                    Assomption = :Assomption => x -> count_pcp_hours(x),
                    Trudeau = :Trudeau => x -> count_pcp_hours(x),
                    StHubert = :StHubert => x -> count_pcp_hours(x))

first(pcp_hours ,5)
````

#### Calcul de la plus grande durée continue de pluie journalière

On cherche ici le maximum d'heures de pluie consécutives dans une journée.

````julia
function longest_continuous_pcp(itr)
    longest = 0
    currentlongest = 0
    israining = false
    for x in itr
        if ismissing(x) || x == 0
            if israining
                israining = false
                currentlongest = 0
                longest = max(longest, currentlongest)
            end
        else
            if israining
                currentlongest += 1
            else
                israining = true
                currentlongest += 1
            end
        end
    end
    max(longest, currentlongest)
end

pcp_continuous_hours = by(data, :date,
                    McTavish = :McTavish => x -> longest_continuous_pcp(x),
                    Bellevue = :Bellevue => x -> longest_continuous_pcp(x),
                    Assomption = :Assomption => x -> longest_continuous_pcp(x),
                    Trudeau = :Trudeau => x -> longest_continuous_pcp(x),
                    StHubert = :StHubert => x -> longest_continuous_pcp(x))

first(pcp_continuous_hours, 5)
````

#### Calcul de la quantité de précipitations par blocs de 3 jours

On additionne les précipitations des deux derniers jours aux précipitations de la journée courante. Nous avons choisi arbitrairement 3 jours, car c'est avec 3 jours qu'il y avait le plus de démarcation entre surverse ou non.

````julia
pcp_3day_sum = DataFrame(pcp_sum)
for i in size(pcp_3day_sum, 1):-1:3
    pcp_3day_sum[i, :McTavish] = sum(skipmissing(pcp_3day_sum[i-2:i, :McTavish]))
    pcp_3day_sum[i, :Bellevue] = sum(skipmissing(pcp_3day_sum[i-2:i, :Bellevue]))
    pcp_3day_sum[i, :Assomption] = sum(skipmissing(pcp_3day_sum[i-2:i, :Assomption]))
    pcp_3day_sum[i, :Trudeau] = sum(skipmissing(pcp_3day_sum[i-2:i, :Trudeau]))
    pcp_3day_sum[i, :StHubert] = sum(skipmissing(pcp_3day_sum[i-2:i, :StHubert]))
end
pcp_3day_sum = pcp_3day_sum[3:end, :]
last(pcp_3day_sum, 5)
````

#### Calcul du maximum de précipitations en 2 heures

On a cherché à savoir la quantité de pluie maximum qui est tombée durant deux heures consécutives.

````julia
function max_2h(itr)
    max_pcp = 0
    for i in 2:size(itr, 1)
        max_pcp = max(max_pcp, itr[i - 1] + itr[i])
    end
    return max_pcp
end

pcp_2h_max = by(data, :date,
                    McTavish = :McTavish => x -> max_2h(x),
                    Bellevue = :Bellevue => x -> max_2h(x),
                    Assomption = :Assomption => x -> max_2h(x),
                    Trudeau = :Trudeau => x -> max_2h(x),
                    StHubert = :StHubert => x -> max_2h(x))

first(pcp_2h_max, 5)
````

#### Calcul du maximum de précipitations en 3 heures

Comme pour la variable explicative précédante, on a cherché à savoir la quantité de pluie maximum qui est tombée durant trois heures consécutives.

````julia
function max_3h(itr)
    max_pcp = 0
    for i in 3:size(itr, 1)
        max_pcp = max(max_pcp, itr[i - 2] + itr[i - 1] + itr[i])
    end
    return max_pcp
end

pcp_3h_max = by(data, :date,
                    McTavish = :McTavish => x -> max_3h(x),
                    Bellevue = :Bellevue => x -> max_3h(x),
                    Assomption = :Assomption => x -> max_3h(x),
                    Trudeau = :Trudeau => x -> max_3h(x),
                    StHubert = :StHubert => x -> max_3h(x))

first(pcp_3h_max, 5)
````

#### Inclusion dans un dataframe des huit variables explicatives potentielles

````julia
ouvrage = "4350-01D"

df = filter(row -> row.NO_OUVRAGE == ouvrage, surverse_df)

x₁ = Array{Union{Missing, Int64}}(missing, size(df, 1)) # variable pour la somme journalière
x₂ = Array{Union{Missing, Int64}}(missing, size(df, 1)) # variable pour le max journalier
x₃ = Array{Union{Missing, Float64}}(missing, size(df, 1)) # variable pour l'intensité journalière
x₄ = Array{Union{Missing, Int64}}(missing, size(df, 1)) # variable pour le nombre d'heures de pluies journalier
x₅ = Array{Union{Missing, Int64}}(missing, size(df, 1)) # variable pour la plus grande durée continue de pluie journalière
x₆ = Array{Union{Missing, Int64}}(missing, size(df, 1)) # variable pour la somme par blocs de trois jours
x₇ = Array{Union{Missing, Int64}}(missing, size(df, 1)) # variable pour le max en 2 heures
x₈ = Array{Union{Missing, Int64}}(missing, size(df, 1)) # variable pour le max en 3 heures


for i=1:size(df,1)
    
    ind = findfirst(pcp_sum[:, :date] .== df[i,:DATE])
    
    x₁[i] = pcp_sum[ind, :McTavish]
    
    ind = findfirst(pcp_max[:, :date] .== df[i, :DATE])
    
    x₂[i] = pcp_max[ind, :McTavish]
    
    ind = findfirst(pcp_intensity[:, :date] .== df[i, :DATE])
     
    x₃[i] = pcp_intensity[ind, :McTavish]
    
    ind = findfirst(pcp_hours[:, :date] .== df[i, :DATE])
     
    x₄[i] = pcp_hours[ind, :McTavish]
    
    ind = findfirst(pcp_continuous_hours[:, :date] .== df[i, :DATE])
     
    x₅[i] = pcp_continuous_hours[ind, :McTavish]
    
    ind = findfirst(pcp_2h_max[:, :date] .== df[i, :DATE])
     
    x₇[i] = pcp_2h_max[ind, :McTavish]
    
    ind = findfirst(pcp_3h_max[:, :date] .== df[i, :DATE])
     
    x₈[i] = pcp_3h_max[ind, :McTavish]
    
end

for i=3:size(df, 1) # on néglige les deux premières dates
   ind = findfirst(pcp_3day_sum[:, :date] .== df[i, :DATE])
     
    x₆[i] = pcp_3day_sum[ind, :McTavish] 
end

df[!, :SUM] = x₁
df[!, :MAX] = x₂
df[!, :INTENSITY] = x₃
df[!, :N_HOURS] = x₄
df[!, :MAX_CONT_HOURS] = x₅
df[!, :THREE_DAY_SUM] = x₆
df[!, :MAX_2H] = x₇
df[!, :MAX_3H] = x₈

variable_columns = [:SUM, :MAX, :INTENSITY, :N_HOURS, :MAX_CONT_HOURS, :THREE_DAY_SUM, :MAX_2H, :MAX_3H]
dropmissing!(df, variable_columns, disallowmissing=true)

first(df, 5)
````

## 3.1 Forêts aléatoires
Nous avons choisi de commencer par utiliser les forêts aléatoires, car c'est une méthode de classification relativement facile à implémenter. En effet, cette méthode demande moins de pré-traitement des données, car nous n'avons pas à nous soucier de problèmes comme la multicolinéarité. Nous avons utilisé l'implémentation des forêts aléatoires de la librairie Julia DecisionTree.jl.

#### Extraction des variables explicatives pour chaque station
Nous avons choisi de considérer toutes les variables explicatives précédemment mentionnées sauf le maximum d'heures continues de précipitation, qui ne semblait pas avoir d'effet sur les surverses. Nous avons donc les variables explicatives suivantes:
- la somme journalière des précipitations
- le taux horaire maximum journalier de précipitations
- l'intensité journalière des précipitations (somme / nombre d'heures de précipitation)
- le nombre d'heures de précipitation journalier
- la somme des précipitations par blocs de trois jours
- le maximum journalier de précipitations en deux heures
- le maximum journalier de précipitations en trois heures

#### Fonctions pour calculer le score f1 par validation croisée
Nous avons créé une fonction qui retourne le score f1 en fonction de la matrice de confusion (vrais positifs, faux positifs, etc.) étant donnée que notre score Kaggle est basé sur cette mesure. Nous avons ensuite modifié une des fonctions de la librairie DecisionTree.jl qui mesure la précision par validation croisée k-folds afin qu'elle utilise le score f1. Cela nous permet donc de comparer la performance de forêts avec différents paramètres. Fonction originale: https://github.com/bensadeghi/DecisionTree.jl/blob/899414020871133fb4189f64e0037cf5c83e1bc1/src/measures.jl#L75

````julia
# calcule le score f1 avec la matrice de confusion
function f1(confusionmatrix::Matrix{Int})
    if size(confusionmatrix) == (1, 1)
        return 1
    end
    
    tp = confusionmatrix[1, 1]
    fp = confusionmatrix[1, 2]
    fn = confusionmatrix[2, 1]
    tn = confusionmatrix[2, 2]
    
    if tp == 0
        return 0
    end
    
    return (2 * tp) / ((2 * tp) + fp + fn)
end
````

````julia
# calcule les k scores f1 par validation croisée k-folds
# originale: https://github.com/bensadeghi/DecisionTree.jl/blob/899414020871133fb4189f64e0037cf5c83e1bc1/src/measures.jl#L75
function _kfoldCV(labels::Vector{T}, features::Matrix{S}, args...; rng) where {S, T}
    mk_rng(rng::Random.AbstractRNG) = rng
    rng = mk_rng(rng)::Random.AbstractRNG
    nfolds = args[1]
    if nfolds < 2
        throw("number of folds must be greater than 1")
    end
    
    n_subfeatures       = args[2]
    n_trees             = args[3]
    partial_sampling    = args[4]
    max_depth           = args[5]
    min_samples_leaf    = args[6]
    min_samples_split   = args[7]
    min_purity_increase = args[8]
    
    N = length(labels)
    ntest = floor(Int, N / nfolds)
    inds = Random.randperm(rng, N)
    accuracy = zeros(nfolds)
    f1scores = zeros(nfolds)
    for i in 1:nfolds
        test_inds = falses(N)
        test_inds[(i - 1) * ntest + 1 : i * ntest] .= true
        train_inds = (!).(test_inds)
        test_features = features[inds[test_inds],:]
        test_labels = labels[inds[test_inds]]
        train_features = features[inds[train_inds],:]
        train_labels = labels[inds[train_inds]]
        
        model = build_forest(
                    labels, features,
                    n_subfeatures,
                    n_trees,
                    partial_sampling,
                    max_depth,
                    min_samples_leaf,
                    min_samples_split,
                    min_purity_increase;
                    rng = rng)
        predictions = apply_forest(model, test_features)
        
        cm = confusion_matrix(test_labels, predictions)
        accuracy[i] = cm.accuracy
        f1scores[i] = f1(cm.matrix)
    end
    println("Mean f1: ", mean(f1scores))
    return f1scores
end

# interface pour _kfoldCV
function kfoldCV(
        labels              :: Vector{T},
        features            :: Matrix{S},
        n_folds             :: Integer,
        n_subfeatures       :: Integer = -1,
        n_trees             :: Integer = 10,
        partial_sampling    :: Float64 = 0.7,
        max_depth           :: Integer = -1,
        min_samples_leaf    :: Integer = 1,
        min_samples_split   :: Integer = 2,
        min_purity_increase :: Float64 = 0.0;
        rng                 = Random.GLOBAL_RNG) where {S, T}
    _kfoldCV(labels, features, n_folds, n_subfeatures, n_trees, partial_sampling,
                max_depth, min_samples_leaf, min_samples_split, min_purity_increase; rng=rng)
end
````

#### Recherche exhaustive d'hyperparamètres optimaux
Afin de trouver les meilleurs paramètres pour chaque forêt aléatoire, nous utilisons la méthode grid search. Celle-ci consiste tout simplement a faire une recherche exhaustive des combinaisons de paramètres parmis une liste de cas possibles. Nous utilisons notre fonction kfoldCV pour déterminer le meilleur modèle.

Les paramètres que nous explorons sont les suivants:
* profondeur maximale pour un arbre: 10, 20, 30 ou pas de maximum
* le nombre minimal de variables explicatives pour être à une feuille: 1, 2, 5 ou 10
* le nombre minimal de variables explicatives pour « brancher » un noeud: 2, 5, 10 ou 15

Les paramètres suivants restent constants:
* partitionnement pour k-folds: 5
* nombre d'arbre par forêt: 1000
* échantillonage partiel: 0.7
* nombre de variables explicatives à considérer pour « branchement »: racine du nombre de variables explicatives

````julia
function gridsearchCV(labels, features)

    k_folds = 5
    n_trees = 1000
    n_subfeatures = -1
    partial_sampling = 0.7
    best_params = Dict(:max_depth => -1, :min_samples_leaf => 5, :min_samples_split => 2)
    best_score = -1

    for max_depth in [-1, 10, 20, 30]
        for min_samples_leaf in [1, 2, 5, 10]
            for min_samples_split in [2, 5, 10, 15]
                score = mean(kfoldCV(labels,
                                features,
                                k_folds,
                                n_subfeatures,
                                n_trees,
                                partial_sampling,
                                max_depth,
                                min_samples_leaf,
                                min_samples_split
                            )
                )
                if score > best_score
                    best_score = score
                    best_params = Dict(:max_depth => max_depth,
                        :min_samples_leaf => min_samples_leaf,
                        :min_samples_split => min_samples_split
                    )
                end
            end
        end
    end
    
    return (best_score, best_params)
    
end
````

## 3.2 Régression logistique avec recherche aléatoire de modèle
Nous avons commencé par faire une régression logistique en utilisant toutes nos variables explicatives. Cependant, le modèle résultant n'était pas très bon. En effet, pour les régressions, contrairement aux forêts aléatoires, nous devons nous soucier de problèmes comme la multicolinéarité et le fait d'avoir trop de variables explicatives peut nuire à notre modèle.

Étant donné notre grand nombre de variables aléatoires, il ne nous était possible de faire une recherche de modèle exhaustive en calculant le BIC pour chaque modèle possible (plus de 2^35 modèles). Nous avons aussi essayé de faire une recherhe stochastique de modèle, mais nous avons rencontré des problèmes computationnels quand nous étions incapables de calculer l'exponentielle du BIC qui est utilisée pour sélectionner le prochain modèle. En effet, nous obtenions des valeurs de BIC très élevées (de 1300 à 1600) et l'exponentielle de ces valeurs en Julia nous donnait la valeur inf, rendant ainsi la recherche impossible.

Suite à cette impasse, nous avons décidé de créer une fonction qui génère un certain nombre de vecteurs aléatoires de variables explicatives à considérer et de calculer le score f1 avec une échantillon d'entrainement et de test. Cette fonction retourne ensuite le meilleur modèle trouvé.

Nous avons utilisé la librairie Julia GLM pour implémenter la régression logistique.

#### Fonction pour avoir le meilleur modèle de régression logistique

```julia
function get_best_prediction(train_data, test_data, train_results, test_results, n_iter)

    bestF1 = 0
    bestArray = []
    bestPrediction = []
    n = length(test_results)
    best_i = 0
    for i = 1:n_iter
        testArray = Set(Int64[])
        nbrElemRem = rand(1:size(train_data, 2) - 1)
        
        sample_features = train_data
        sample_tests = test_data
        
        for j=1: nbrElemRem
            x = rand(1:size(sample_features, 2))
            push!(testArray, rand(1:size(sample_features, 2)))
        end

        sample_features = sample_features[:, setdiff(1:end, testArray)]
        sample_tests = sample_tests[:, setdiff(1:end, testArray)]

        try
            M = glm(sample_features, train_results, Bernoulli(), LogitLink())
            θ̂ = GLM.predict(M, sample_tests)       
            for i = 0.1:0.1:0.9
                Ŷ = zeros(Int64,n)
                Ŷ[θ̂.>i] .= 1;
                f1 = f1_func(test_results, Ŷ)
                if f1 > bestF1
                    bestF1 = f1
                    best_i = i
                    bestArray = testArray
                    bestPrediction = Ŷ
                end
            end
        catch e
            print("Problème")
        end
    end
    return (bestArray)
end
```

### Performance du modèle
Avec ce modèle, nous avons réussi à atteindre un score d'environ 0.67 sur Kaggle, ce qui est significativement plus bas que notre modèle qui utilise les forêts aléatoires. Nous n'avons donc pas sélectionné ce modèle comme modèle final.

```julia
function segmentdata(value, min, max)
    if value < ((1 / 3) * (max - min)) + min
        return [1, 0, 0]
    elseif value < ((2 / 3) * (max - min)) + min
        return [0, 1, 0]
    else
        return [0, 0, 1]
    end
end

function segmentdataframe(df, dfmin, dfmax)
    df_segmented = DataFrame(date = Date[],
        McTavish_LOW = Int[], McTavish_MED = Int[], McTavish_HIGH = Int[],
        Bellevue_LOW = Int[], Bellevue_MED = Int[], Bellevue_HIGH = Int[],
        Assomption_LOW = Int[], Assomption_MED = Int[], Assomption_HIGH = Int[],
        Trudeau_LOW = Int[], Trudeau_MED = Int[], Trudeau_HIGH = Int[],
        StHubert_LOW = Int[], StHubert_MED = Int[], StHubert_HIGH = Int[])

    for i in 1:size(df, 1)

        Bellevue = segmentdata(df[i, :Bellevue], dfmin, dfmax)

    end
    
    return df_segmented
end
```

```julia
function f1_func(y, y_pred)
    true_positives = sum((y .== 1) .& (y_pred .== 1))
    false_positives = sum((y .== 0) .& (y_pred .== 1))
    false_negatives = sum((y .== 1) .& (y_pred .== 0))
    precision = true_positives / (true_positives + false_positives)
    recall = true_positives / (true_positives + false_negatives)
    if isnan(recall) || isnan(precision)
        return 0
    end
    return 2 * ((precision * recall) / (precision + recall))
end
```

```julia
    for col in 4:size(df, 2)
```

## 3.4 Régression logistique avec analyse en composantes principales

Cet essai sur la regression logistique visait a utiliser seulement les variables aléatoires qui améliorait le plus la regression logistique

``` Julia
function reduce_func(new_features, new_labels,new_features_test,new_labels_test,percent)
    for i=1:size(features_3260_01D)[2]
            lessfeatures = new_features[:,1:end .!= i]
        
            lessfeatures_test = new_features_test[:,1:end .!= i]
            
            M = glm(lessfeatures, new_labels, Bernoulli(), LogitLink())
            n₁ = length(new_labels_test)
```

#### Prédiction des surverses pour le fichier test
Selon l'ouvrage à prédire, on utilise la bonne forêt. On sauvegarde ensuite le dataframe dans un fichier csv pour la soumission.

````julia
submission = DataFrame(ID = String[], Surverse = Int[])

for i in 1:size(df, 1)
    no_ouvrage = df[i, :NO_OUVRAGE]
    date = df[i, :DATE]
    surverse = 0
    j = findfirst((df.DATE .== date) .& (df.NO_OUVRAGE .== no_ouvrage))
    if no_ouvrage == "3260-01D"
        surverse = apply_forest(model_3260_01D, convert(Vector{Float64}, select(df, Not([:NO_OUVRAGE, :DATE]))[j, :]))
    elseif no_ouvrage == "3350-07D"
        surverse = apply_forest(model_3350_07D, convert(Vector{Float64}, select(df, Not([:NO_OUVRAGE, :DATE]))[j, :]))
    elseif no_ouvrage == "4350-01D"
        surverse = apply_forest(model_4240_01D, convert(Vector{Float64}, select(df, Not([:NO_OUVRAGE, :DATE]))[j, :]))
    elseif no_ouvrage == "3260-01D"
        surverse = apply_forest(model_4350_01D, convert(Vector{Float64}, select(df, Not([:NO_OUVRAGE, :DATE]))[j, :]))
    elseif no_ouvrage == "4380-01D"
        surverse = apply_forest(model_4380_01D, convert(Vector{Float64}, select(df, Not([:NO_OUVRAGE, :DATE]))[j, :]))
    end
    push!(submission, (no_ouvrage * "_" * string(date), Int(surverse)))
end

CSV.write("submission.csv", submission)
````
